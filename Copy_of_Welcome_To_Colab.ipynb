```
def is_Prime(n):
  if n < 2:
    return False
  for i in range(2, (n - 1)):
    if n % i == 0:
      return False
  return True
n = int(input("Nh·∫≠p v√†o gi√° tr·ªã N: "))
if is_Prime(n):
  print(f"{n} l√† s·ªë nguy√™n t·ªë!")
else:
  print(f"{n} kh√¥ng ph·∫£i l√† s·ªë nguy√™n t·ªë!")
```

```
def tien_thue(tong_thu_nhap, nguoi_phu_thuoc):
  thue_ca_nhan = 11
  thue_phu_thuoc = 4.4 * nguoi_phu_thuoc
  chiu_thue = tong_thu_nhap - (thue_ca_nhan + thue_phu_thuoc)
  thue_phai_dong = 0
  tmp = chiu_thue
  if chiu_thue < 0:
    return 0
  if tmp > 80:
    thue_phai_dong += (tmp - 80) * 0.35
    tmp = 80
  if tmp > 52:
    thue_phai_dong += (tmp - 52) * 0.30
    tmp = 52
  if tmp > 32:
    thue_phai_dong += (tmp - 32) * 0.25
    tmp = 32
  if tmp > 18:
    thue_phai_dong += (tmp - 18) * 0.20
    tmp = 18
  if tmp > 10:
    thue_phai_dong += (tmp - 10) * 0.15
    tmp = 10
  if tmp > 5:
    thue_phai_dong += (tmp - 5) * 0.10
    tmp = 5
  if tmp > 0:
    thue_phai_dong += (tmp * 0.05)
  return thue_phai_dong
tong_thu_nhap = float(input("Nh·∫≠p v√†o t·ªïng thu nh·∫≠p: "))
nguoi_phu_thuoc = int(input("Nh·∫≠p v√†o s·ªë ng∆∞·ªùi ph·ª• thu·ªôc: "))
tien_thue(tong_thu_nhap, nguoi_phu_thuoc)
```

```
import math
def PT2(a, b, c, delta):
  if delta > 0:
    return (-b + math.sqrt(delta)) / 2 * a, (-b - math.sqrt(delta)) / 2 * a
  if delta < 0:
    return "Phuong trinh vo nghiem!"
  if delta == 0:
    return -b / 2 * a
a = int(input("Nhap vao gia tri A: "))
b = int(input("Nhap vao gia tri B: "))
c = int(input("Nhap vao gia tri C: "))
delta = b * b - 4 * a * c
print(PT2(a, b, c, delta))
```

```
def tbc(n):
    avg = 0
    dem = 0
    while n > 0:
        so = n % 10
        avg += so
        dem += 1
        n //= 10
    return avg / dem
n = int(input())
print(tbc(n))
```

```

import math
class PhanSo:
  tu = 0
  mau = 1
  def __init__(self, tu, mau):
    if mau == 0:
      raise ValueError("M·∫´u s·ªë kh√¥ng th·ªÉ b·∫±ng 0!")
    self.tu = tu
    self.mau = mau
    #self.rutgon()
  def rutgon(self):
    ucln = math.gcd(self.tu, self.mau)
    self.tu //= ucln
    self.mau //= ucln
    return True
  def nhap(self):
    self.tu = int(input("Nh·∫≠p v√†o t·ª≠ s·ªë: "))
    self.mau = int(input("Nh·∫≠p v√†o m·∫´u s·ªë: "))
  def xuat(self):
    print(f"{self.tu} / {self.mau}")
  def doi_dau(self):
    if self.mau < 0:
      self.tu *= -1
      self.mau *= -1
  def nghich_dao(self):
    if self.tu == 0:
      return False
    tmp = self.tu
    self.tu = self.mau
    self.mau = tmp
  def la_phan_so_toi_gian(self, other):
    uc1 = math.gcd(self.tu, self.mau)
    if uc1 >= 2:
        if self.tu % uc1 == 0 and self.mau % uc1 == 0:
            print(f"{self.tu} / {self.mau} kh√¥ng l√† ph√¢n s·ªë t·ªëi gi·∫£n")
    else:
      print(f"{self.tu} / {self.mau} l√† ph√¢n s·ªë t·ªëi gi·∫£n")
    uc2 = math.gcd(other.tu, other.mau)
    if uc2 >= 2:
        if other.tu % uc2 == 0 and other.mau % uc2 == 0:
            print(f"{other.tu} / {other.mau} kh√¥ng l√† ph√¢n s·ªë t·ªëi gi·∫£n")
    else:
        print(f"{other.tu} / {other.mau} l√† ph√¢n s·ªë t·ªëi gi·∫£n")
  def la_phan_so_duong(self, other):
    if self.tu > 0 and self.mau > 0:
      print(f"{self.tu} / {self.mau} L√† ph√¢n s·ªë d∆∞∆°ng!")
    else:
      print(f"{self.tu} / {self.mau} L√† ph√¢n s·ªë √¢m!")
    if other.tu > 0 and other.mau > 0:
        print(f"{other.tu} / {other.mau} L√† ph√¢n s·ªë d∆∞∆°ng!")
    else:
        print(f"{other.tu} / {other.mau} L√† ph√¢n s·ªë √¢m!")
  def __eq__(self, other):
    if self.tu == other.tu and self.mau == other.mau:
      print("Hai ph√¢n s·ªë b·∫±ng nhau!")
    else:
      print("Hai ph√¢n s·ªë kh√¥ng b·∫±ng nhau!")
  def __gt__(self, other):
    ps1 = self.tu / self.mau
    ps2 = other.tu / other.mau
    if ps1 > ps2:
      print(f"{self.tu} / {self.mau} > {other.tu} / {other.mau}")
    else:
      print(f"{self.tu} / {self.mau} < {other.tu} / {other.mau}")
  def gia_tri_thap_phan(self):
    return self.tu / self.mau
  def la_phan_so_nguyen(self):
    self.tu //= self.mau
    ps = self.tu
    if ps:
      print(f"{ps} Ph√¢n s·ªë l√† s·ªë nguy√™n")
    else:
      print(f"{ps} Ph√¢n s·ªë kh√¥ng l√† s·ªë nguy√™n")
  def doi_sang_so_thap_phan(self, k):
    ps = self.tu / self.mau
    return format(ps, f'.{k}f')

ps1 = PhanSo(0, 1)
ps2 = PhanSo(0, 1)
ps1.nhap()
ps2.nhap()
while True:
    chon = int(input("Chon bai: "))
    match(chon):
        case 0:
            break
        case 1:
            ps1.xuat()
            ps2.xuat()
        case 2:
            ps1.rutgon()
            ps2.rutgon()
            ps1.xuat()
            ps2.xuat()
        case 3:
            ps1.nghich_dao()
            ps1.xuat()
        case 4:
            ps1.la_phan_so_toi_gian(ps2)
        case 5:
            ps1.la_phan_so_duong(ps2)
        case 6:
            ps2 = PhanSo(0, 1)
            ps2.nhap()
            ps2.xuat()
            ps1.__eq__(ps2)
        case 7:
            ps2 = PhanSo(0, 1)
            ps2.nhap()
            ps2.xuat()
            ps1.__gt__(ps2)
        case 8:
            ps1.doi_dau()
            ps1.xuat()
        case 9:
            print(ps1.gia_tri_thap_phan())
            print(ps2.gia_tri_thap_phan())
        case 10:
            ps1.la_phan_so_nguyen()
        case 11:
            k = int(input("Nhap vao so k: "))
            print(ps1.doi_sang_so_thap_phan(k))
```

```
import math
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def __str__(self):
    return self.x, self.y
  def is_on_x_axis(self):
    if self.y == 0:
      return True
    else:
      return False
  def is_on_y_axis(self):
    if self.x == 0:
      return True
    else:
      return False
  def is_in_first_quadrant(self):
    if self.x > 0 and self.y > 0:
      return True
    return False
  def is_in_second_quadrant(self):
    if self.x < 0 and self.y > 0:
      return True
    return False
  def is_in_third_quadrant(self):
    if self.x < 0 and self.y < 0:
      return True
    return False
  def is_in_fourth_quadrant(self):
    if self.x > 0 and self.y < 0:
      return True
    return False
  def is_symmetric_x_axis(self, other):
    if self.x == other.x and self.y == -other.y:
      return True
    return False
  def is_symmetric_y_axis(self, other):
    if self.y == other.y and self.x == -other.x:
      return True
    return False
  def is_symmetric_origin(self, other):
    if self.x == -other.x and self.y == -other.y:
      return True
    return False
  def is_origin(self):
    if self.x == 0 and self.y == 0:
      return True
    return False
  def distance_to(self, other):
    return math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)
  def nhap_toa_do(self):
    self.x = int(input("Nhap vao X: "))
    self.y = int(input("Nhap vao Y: "))
  def xuat_toa_do(self):
    print(f"{self.x}, {self.y}")
pt = Point(1, 0)
pt.nhap_toa_do()

while True:
  chon = int(input("Chon bai: "))
  match(chon):
    case 0:
      break
    case 1:
      pt.xuat_toa_do()
    case 2:
      print(pt.__str__())
    case 3:
      print(pt.is_on_x_axis())
    case 4:
      print(pt.is_on_y_axis())
    case 5:
      print(pt.is_in_first_quadrant())
    case 6:
      print(pt.is_in_second_quadrant())
    case 7:
      print(pt.is_in_third_quadrant())
    case 8:
      print(pt.is_in_fourth_quadrant())
    case 9:
      pt2 = Point(0, 0)
      pt2.nhap_toa_do()
      pt2.xuat_toa_do()
      print(pt.is_symmetric_x_axis(pt2))
    case 10:
      pt2 = Point(0, 0)
      pt2.nhap_toa_do()
      pt2.xuat_toa_do()
      print(pt.is_symmetric_y_axis(pt2))
    case 11:
      pt2 = Point(0, 0)
      pt2.nhap_toa_do()
      pt2.xuat_toa_do()
      print(pt.is_symmetric_origin(pt2))
    case 12:
      print(pt.is_origin())
    case 13:
      pt2 = Point(0, 0)
      pt2.nhap_toa_do()
      pt2.xuat_toa_do()
      print(pt.distance_to(pt2))

```

```
import math
class TamGiac:
  def __init__(self, x1, x2, x3, y1, y2, y3):
    self.x1 = x1
    self.x2 = x2
    self.x3 = x3
    self.y1 = y1
    self.y2 = y2
    self.y3 = y3
  def nhap_diem(self):
    self.x1 = int(input("Nhap vao X1: "))
    self.x2 = int(input("Nhap vao X2: "))
    self.x3 = int(input("Nhap vao X3: "))
    self.y1 = int(input("Nhap vao y1: "))
    self.y2 = int(input("Nhap vao y2: "))
    self.y3 = int(input("Nhap vao y3: "))
  def xuat_diem(self):
    print(f"({self.x1}, {self.y1}), ({self.x2}, {self.y2}), ({self.x3}, {self.y3})")
  def trung_diem_tam_giac(self):
    M1 = (self.x1 + self.x2) / 2
    M2 = (self.y1 + self.y2) / 2
    N1 = (self.x2 + self.x3) / 2
    N2 = (self.y2 + self.y3) / 2
    P1 = (self.x3 + self.x1) / 2
    P2 = (self.y3 + self.y1) / 2
    print(f"M = ({M1}, {M2})")
    print(f"N = ({N1}, {N2})")
    print(f"P = ({P1}, {P2})")
    return (M1, M2), (N1, N2), (P1, P2)
  def toa_do_trong_tam(self):
    G1 = (self.x1 + self.x2 + self.x3) / 3
    G2 = (self.y1 + self.y2 + self.y3) / 3
    print(f"G = ({G1:.2f}, {G2:.2f}) ")
  def do_dai_trung_binh(self):
    (M1, M2), (N1, N2), (P1, P2) = self.trung_diem_tam_giac()
    Ma = math.sqrt(((N1 - P1) ** 2) + ((N2 - P2) ** 2))
    Mb = math.sqrt(((P1 - M1) ** 2) + ((P2 - M2) ** 2))
    Mc = math.sqrt(((M1 - N1) ** 2) + ((M2 - N2) ** 2))
    print(f"MA = {Ma:.2f}, MB = {Mb:.2f}, MC = {Mc:.2f}")
tg = TamGiac(0, 0, 0, 0, 0, 0)
tg.nhap_diem()
tg.xuat_diem()
print("Trung diem cua tam giac la: ")
tg.trung_diem_tam_giac()
print("Toa do trong tam cua tam giac: ")
tg.toa_do_trong_tam()
print("Do dai 3 duong trung binh cua tam giac")
tg.do_dai_trung_binh()
```

```
class DongVat:
  def __init__(self, ten, loai, mau_sac):
    self.ten = ten
    self.loai = loai
    self.mau_sac = mau_sac
  def nhapTT(self):
    self.ten = input("Nh·∫≠p v√†o t√™n ƒë·ªông v·∫≠t: ")
    self.mau_sac = input("Nh·∫≠p v√†o m√†u s·∫Øc: ")
  def xuatTT(self):
    print(f"{self.ten} l√† lo√†i {self.loai} v·ªõi m√†u s·∫Øc {self.mau_sac}")
class Meo(DongVat):
  def __init__(self,ten, loai, mau_sac, giong):
    super().__init__(ten, "M√®o", mau_sac)
    self.giong = giong
  def nhapTTMeo(self):
    self.nhapTT()
    self.giong = input("Nh·∫≠p v√†o gi·ªëng m√®o: ")
  def keu(self):
    print("Meo...meo")
  def xuatTTMeo(self):
    print(f"{self.ten} l√† lo√†i {self.loai} v·ªõi gi·ªëng {self.giong} v√† m√†u s·∫Øc {self.mau_sac}")
    self.keu()
class Huou(DongVat):
  def __init__(self, loai_huou, ten, loai, mau_sac):
    super().__init__(ten, "H∆∞∆°u", "")
    self.loai_huou = loai_huou
  def nhapTTHuou(self):
    self.nhapTT()
    self.loai_huou = input("Nh·∫≠p v√†o lo·∫°i h∆∞∆°u: ")
  def keu(self):
    print("Gr·ª´m...gr·ª´m")
  def xuatTTHuou(self):
    print(f"{self.ten} l√† lo√†i {self.loai} v·ªõi lo·∫°i {self.loai_huou}")
    self.keu()
class Chim(DongVat):
  def __init__(self, ten, loai, mau_sac):
    super().__init__(ten, "Chim", mau_sac)
  def keu(self):
    print("L√≠u lo...l√≠u lo")
  def nhapTTChim(self):
    self.nhapTT()
  def xuatTTChim(self):
    self.xuatTT()
    self.keu()
meo = Meo("", "", "", "")
meo.nhapTTMeo()

huou = Huou("", "", "", "")
huou.nhapTTHuou()

chim = Chim("", "", "")
chim.nhapTTChim()

meo.xuatTTMeo()
huou.xuatTTHuou()
chim.xuatTTChim()
```
